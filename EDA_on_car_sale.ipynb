```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Car Sale.csv")  # Update with your actual file path

# Get basic info about the dataset
print(df.info())

# Summary statistics for numeric columns
print(df.describe())
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()

print("Missing values per column:")
print(missing_values)

# Visualizing missing values with a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram of car prices
plt.figure(figsize=(8, 5))
sns.histplot(df["Price ($)"], bins=30, kde=True)

plt.title("Distribution of Car Prices")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Count the number of cars sold by each manufacturer
manufacturer_counts = df["Company"].value_counts()

print("Number of cars sold per manufacturer:")
print(manufacturer_counts)

# Visualizing manufacturer sales
plt.figure(figsize=(10, 5))
sns.barplot(x=manufacturer_counts.index, y=manufacturer_counts.values, palette="viridis")
plt.xticks(rotation=90)
plt.title("Cars Sold per Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for transmission type vs. car price
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Transmission"], y=df["Price ($)"])

plt.title("Transmission Type vs. Car Price")
plt.xlabel("Transmission Type")
plt.ylabel("Price (USD)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count car sales by dealer region
dealer_region_counts = df["Dealer_Region"].value_counts()

print("Car sales distribution by dealer region:")
print(dealer_region_counts)

# Visualizing dealer region sales
plt.figure(figsize=(8, 5))
sns.barplot(x=dealer_region_counts.index, y=dealer_region_counts.values, palette="magma")
plt.title("Car Sales Distribution by Dealer Region")
plt.xlabel("Dealer Region")
plt.ylabel("Number of Cars Sold")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot for Annual Income vs. Car Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Annual Income"], y=df["Price ($)"])

plt.title("Annual Income vs. Car Price")
plt.xlabel("Annual Income (USD)")
plt.ylabel("Car Price (USD)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count car sales by body style
body_style_counts = df["Body Style"].value_counts()

print("Most popular car body styles:")
print(body_style_counts)

# Visualizing body style popularity
plt.figure(figsize=(8, 5))
sns.barplot(x=body_style_counts.index, y=body_style_counts.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Popularity of Car Body Styles")
plt.xlabel("Body Style")
plt.ylabel("Number of Cars Sold")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count car purchases by model
model_counts = df["Model"].value_counts()

print("Most commonly purchased car models:")
print(model_counts)

# Visualizing popular car models
plt.figure(figsize=(14, 12))
sns.barplot(x=model_counts.index, y=model_counts.values, palette="rocket")
plt.xticks(rotation=90)
plt.title("Popularity of Car Models")
plt.xlabel("Car Model")
plt.ylabel("Number of Cars Sold")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count car purchases by gender
gender_counts = df["Gender"].value_counts()

print("Car purchases by gender:")
print(gender_counts)

# Visualizing gender preferences in car purchases
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="pastel")

plt.title("Car Purchases by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Cars Purchased")
plt.show()
```

```python
# Find the highest car price recorded
max_price = df["Price ($)"].max()

# Retrieve the corresponding customer(s)
customer_with_max_price = df[df["Price ($)"] == max_price][["Customer Name", "Company", "Model"]]

print(f"The highest car price recorded is ${max_price}. Here are the customers who purchased them:")
print(customer_with_max_price)
```
